## Quantification of species

```python
# Extract the family name from the 'ssciname' column
df['family'] = df['ssciname'].str.split(',').str[0]

# Group by the family and aggregate the counts
family_counts = df.groupby('family').agg({
    'ssciname': 'count',
    'evalue': 'mean',
    'pident': 'mean',
    'qcovs': 'mean'
}).rename(columns={'ssciname': 'species_count'})

# Sort the results by the species count in descending order
family_counts = family_counts.sort_values(by='species_count', ascending=False)

# Display the results
display(family_counts)
```

```python
# Filter the DataFrame to include only Streptococcus species
streptococcus_df = df[df['ssciname'].str.contains('Streptococcus')]

# Calculate the total number of species and the number of Streptococcus species
total_species_count = len(df)
streptococcus_count = len(streptococcus_df)

# Calculate the percentage of Streptococcus species
percentage_streptococcus = (streptococcus_count / total_species_count) * 100

# Display the results
print(f'Total species count: {total_species_count}')
print(f'Streptococcus species count: {streptococcus_count}')
print(f'Percentage of Streptococcus species: {percentage_streptococcus:.2f}%')
```

## Brachybacterium, Brevundimonas, and Bradyrhizobium were deemed environmental contaminants and are not known to be associated with CNS infectious diseases

```python
# List of families to exclude
exclude_families = ['Brachybacterium', 'Brevundimonas', 'Bradyrhizobium']

# Filter the DataFrame to exclude specified families
filtered_df = df[~df['family'].isin(exclude_families)]

# Filter the new DataFrame to include only Streptococcus species
streptococcus_filtered_df = filtered_df[filtered_df['ssciname'].str.contains('Streptococcus')]

# Calculate the total number of species and the number of Streptococcus species after exclusion
total_species_count_filtered = len(filtered_df)
streptococcus_count_filtered = len(streptococcus_filtered_df)

# Calculate the percentage of Streptococcus species after exclusion
percentage_streptococcus_filtered = (streptococcus_count_filtered / total_species_count_filtered) * 100

# Display the results
print(f'Total species count (after exclusion): {total_species_count_filtered}')
print(f'Streptococcus species count (after exclusion): {streptococcus_count_filtered}')
print(f'Percentage of Streptococcus species (after exclusion): {percentage_streptococcus_filtered:.2f}%')
```

## Unsupervised learning

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Extract relevant columns
features = ['evalue', 'pident', 'qcovs']
X = df[features]

# Log transform evalue to bring it to a similar scale as other features
X.loc[:, 'evalue'] = X['evalue'].apply(lambda x: -1 if x == 0 else x)  # Avoid log(0)
X.loc[:, 'evalue'] = X['evalue'].apply(lambda x: -1 * (abs(x) + 1) if x < 0 else x)  # Avoid log(negative)
X.loc[:, 'evalue'] = X['evalue'].apply(lambda x: abs(x))
X.loc[:, 'evalue'] = X['evalue'].apply(lambda x: 0 if x == 0 else x)  # Log(1) = 0
X.loc[:, 'evalue'] = X['evalue'].apply(lambda x: -1 * x if x != 0 else x)  # Log-transform

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform t-SNE for dimensionality reduction
tsne = TSNE(n_components=2, random_state=42)
X_tsne_first= tsne.fit_transform(X_scaled)

# Visualize t-SNE results
plt.scatter(X_tsne_first[:, 0], X_tsne_first[:, 1], alpha=0.5)
plt.title('t-SNE Visualization')
plt.show()
```

```python
# Perform K-Means clustering for original t-SNE
num_clusters = 4  # adjust the number of clusters
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
X_clustered = kmeans.fit_predict(X_scaled)

# Visualize the clustered data
plt.scatter(X_tsne_first[:, 0], X_tsne_first[:, 1], c=X_clustered, cmap='viridis', alpha=0.5)
plt.title('t-SNE Visualization with K-Means Clustering')
plt.show()

# Add cluster labels to the DataFrame
df['cluster'] = X_clustered

# Display the DataFrame with cluster labels
print(df[['qseqid', 'ssciname', 'evalue', 'pident', 'qcovs', 'cluster']])
```

## Rewrite code according to comments with actual log_transformation

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np

# Extract relevant columns
features = ['evalue', 'pident', 'qcovs']
X = df[features]

# Replace evalues of 0 with the next smallest positive value
min_positive_evalue = np.sort(X['evalue'].replace(0, np.nan).dropna().unique())[0]
X['evalue'] = X['evalue'].replace(0, min_positive_evalue)

# Log transform evalue with base 10
X['evalue'] = -np.log10(X['evalue'])

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### For non-scaled X

```python
# Perform t-SNE for dimensionality reduction (for X)
tsne = TSNE(n_components=2, random_state=42)
X_tsne_log = tsne.fit_transform(X)

# Visualize t-SNE results
plt.scatter(X_tsne_log[:, 0], X_tsne_log[:, 1], alpha=0.5)
plt.title('t-SNE Visualization')
plt.show()
```

### For scaled

```python
# Perform t-SNE for dimensionality reduction (for X_scaled)
tsne = TSNE(n_components=2, random_state=42)
X_sc_tsne_log = tsne.fit_transform(X_scaled)

# Visualize t-SNE results
plt.scatter(X_sc_tsne_log[:, 0], X_sc_tsne_log[:, 1], alpha=0.5)
plt.title('t-SNE Visualization')
plt.show()
```

### Comparison of t-sne for scaled and non-scaled

```python
plt.figure(figsize=(12, 6))

# Plot Log Transformed t-SNE non-scaled
plt.subplot(1, 2, 1)
plt.scatter(X_tsne_log[:, 0], X_tsne_log[:, 1], alpha=0.5)
plt.title('Log Transformed non-scaled t-SNE Visualization')

# Plot scaled t-SNE
plt.subplot(1, 2, 2)
plt.scatter(X_sc_tsne_log[:, 0], X_sc_tsne_log[:, 1], alpha=0.5)
plt.title('Log transformed scaled t-SNE Visualization')

plt.tight_layout()
plt.show()
```

### Performing K-means for scaled and non-scaled t-sne

```python
from sklearn.cluster import KMeans

# Number of clusters
num_clusters = 6  # Adjust the number of clusters as needed

# Perform K-Means clustering for non-scaled t-SNE
kmeans_original = KMeans(n_clusters=num_clusters, random_state=42)
X_clustered_original = kmeans_original.fit_predict(X_tsne_log)

# Perform K-Means clustering for scaled t-SNE
kmeans_scaled = KMeans(n_clusters=num_clusters, random_state=42)
X_clustered_scaled = kmeans_scaled.fit_predict(X_sc_tsne_log)

# Add cluster labels to the DataFrame
df['cluster_original'] = X_clustered_original
df['cluster_scaled'] = X_clustered_scaled

# Visualize the clustered data for non-scaled t-SNE
plt.scatter(X_tsne_log[:, 0], X_tsne_log[:, 1], c=X_clustered_original, cmap='viridis', alpha=0.5)
plt.title('t-SNE Visualization with K-Means Clustering (Non-Scaled)')
plt.show()

# Visualize the clustered data for scaled t-SNE
plt.scatter(X_sc_tsne_log[:, 0], X_sc_tsne_log[:, 1], c=X_clustered_scaled, cmap='viridis', alpha=0.5)
plt.title('t-SNE Visualization with K-Means Clustering (Scaled)')
plt.show()

# Display the DataFrame with cluster labels
print(df[['qseqid', 'ssciname', 'evalue', 'pident', 'qcovs', 'cluster_original', 'cluster_scaled']])
```
